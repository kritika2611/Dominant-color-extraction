```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('image2.jpg')
```

```python
img = img[:,250:650,:]
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
img.shape
```

```python
X = img.reshape((-1,3))
print(X.shape)
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=5)
km.fit(X)
```

```python
centers = km.cluster_centers_
```

```python
print(centers)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for c in centers:
    ax.scatter(c[0], c[1], c[2], c='b', marker='*')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```

```python
import numpy as np
colors = []
for i,color in enumerate(centers):
    plt.subplot(1,10,i+1)
    colors.append(color)

    square = np.zeros((100,100,3),dtype='uint8')
    square[:,:,:] = color
    plt.imshow(square)
    plt.axis("off")
    
plt.show()
    
```

```python
np.unique(km.labels_,return_counts=True)
```

```python
# Repaint the Original Image
labels = km.labels_
```

```python
new_img  = np.zeros(X.shape)
for ix in range(X.shape[0]):
    new_img[ix] = colors[labels[ix]]

print(new_img.shape)
```

```python
new_img = new_img.reshape(img.shape)
```

```python
#refilling image
new_img = new_img/255
plt.imshow(np.hstack((img/255,np.ones((445,10,3)),new_img)))
plt.axis("off")
plt.title("Before and After :)")
plt.show()
```

## REPAINTING USING 10 COLORS

```python
img1 = cv2.imread('image1.jpg')
```

```python
img1 = img1[:,250:650,:]
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.axis("off")
plt.show()
```

```python
img1.shape
```

```python
X = img1.reshape((-1,3))
print(X.shape)
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=10)
km.fit(X)
```

```python
centers = km.cluster_centers_
```

```python
print(centers)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for c in centers:
    ax.scatter(c[0], c[1], c[2], c='r', marker='*',s=25)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```

```python
colors = []
for i,color in enumerate(centers):
    plt.subplot(1,10,i+1)
    colors.append(color)

    square = np.zeros((100,100,3),dtype='uint8')
    square[:,:,:] = color
    plt.imshow(square)
    plt.axis("off")
    
plt.show()
```

```python
# Repaint the Original Image
labels = km.labels_
```

```python
new_img  = np.zeros(X.shape)
for ix in range(X.shape[0]):
    new_img[ix] = colors[labels[ix]]

print(new_img.shape)
```
